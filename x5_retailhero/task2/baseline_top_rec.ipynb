```
clients = pd.read_csv(os.path.join(PATH_TO_FILE, './data/raw/clients.csv'),
                      parse_dates=['first_issue_date', 'first_redeem_date'])
clients.head(3)
```

```
purch = pd.read_csv(os.path.join(PATH_TO_FILE, './data/raw/purchases.csv'),
                    index_col=0,
                    parse_dates=['transaction_datetime'])
```

```
for feat in purch.columns:
  if str(purch[feat].dtype).startswith('float64'):
    purch[feat] = pd.to_numeric(purch[feat], downcast='float')

purch.info()
```

```
purch.head(3)
```

```
purch = purch.merge(clients[['client_id', 'gender']], on='client_id')
purch.sample(10)
```

```
class TopRecommender:
  def __init__(self, gender_split=False):
    self._rec_products = []
    self._gender_rec_products = {}
    self._gender_split = gender_split
  
  @property
  def rec_products(self):
    return self._rec_products

  @property
  def gender_split(self):
    return self._gender_split

  @property
  def gender_rec_products(self):
    return self._gender_rec_products

  def train(self):
    if self.gender_split:
      self.gender_rec_products['M'] = purch[purch['gender'] == 'M']\
                          ['product_id'].value_counts().\
                          sort_values(ascending=False).index
      self.gender_rec_products['F'] = purch[purch['gender'] == 'F']\
                          ['product_id'].value_counts().\
                          sort_values(ascending=False).index
      self.gender_rec_products['U'] = purch[purch['gender'] == 'U']\
                          ['product_id'].value_counts().\
                          sort_values(ascending=False).index
    else:
      self.rec_products = list(purch['product_id'].value_counts().\
                sort_values(ascending=False).index)

  def predict(self, n_products=30):
    if self.gender_split:
      return self.gender_rec_products
    else:
      return self.rec_products[:n_products]
```

```
top_rec = TopRecommender(gender_split=True)
top_rec.train()
```
